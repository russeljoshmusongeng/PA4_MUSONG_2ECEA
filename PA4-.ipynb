Part 1a

```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Get the data from "board2" and computing the average for the boards since it was not indicated.
df = pd.read_excel("board2.xlsx")
        
df["Average"] = df[["Math", "GEAS", "Electronics", "Communication"]].mean(axis=1)
```

```python
# Create a data frame called Instru, 
# subsetting the columns and indicating who was greater than 70 in electronics
Instru = df[
    (df["Track"] == "Instrumentation") &
    (df["Hometown"] == "Luzon") &
    (df["Electronics"] > 70)
][["Name", "GEAS", "Electronics"]]
```

```python
# Save the output and Printing the output is optional
Instru.to_csv("Instru.csv", index=False)
print("Instru DataFrame")
print(Instru.head())
```

Part 2b

```python
#Create a Data frame called Mindy and same style as the previous 
#but with the calculated average greater than or equal to 55
Mindy = df[
    (df["Hometown"] == "Mindanao") &
    (df["Gender"] == "Female") &
    (df["Average"] >= 55)
][["Name", "Track", "Electronics", "Average"]]
```

```python
# Save the output, printing the output is optional
Mindy.to_csv("Mindy.csv", index=False)
print("Mindy DataFrame:")
print(Mindy.head())
```

Part 2

```python
# Bar graph Average by Track
#Using the given data and inputting needed titles, labels and plotting the data
avg_track = df.groupby("Track")["Average"].mean()
plt.figure()
avg_track.plot(kind="bar")
plt.title("Average Score by Track")
plt.xlabel("Track")
plt.ylabel("Average")
plt.show()
```

```python
# Bar graph Average by Gender
#Using the given data and inputting needed titles, labels and plotting the data
avg_gender = df.groupby("Gender")["Average"].mean()
plt.figure()
avg_gender.plot(kind="bar")
plt.title("Average Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average")
plt.show()
```

```python
#Bar graph Average by Hometown
#Using the given data and inputting needed titles, labels and plotting the data
avg_home = df.groupby("Hometown")["Average"].mean()
plt.figure()
avg_home.plot(kind="bar")
plt.title("Average Score by Hometown")
plt.xlabel("Hometown")
plt.ylabel("Average")
plt.show()
```
